```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset_info.csv')
```

```python
df.head() 
```

```python
# Compute available transcripts
df["Available transcripts"] = df["Recordings Count"] - df["Missing transcripts"]

# Plot stacked bars
plt.figure(figsize=(10, 6))
plt.bar(df["Group"], df["Available transcripts"], label="Available Transcripts", color="#98DE9C")
plt.bar(df["Group"], df["Missing transcripts"], bottom=df["Available transcripts"], label="Missing Transcripts", color="#C7B1B0")

plt.title("Recording Availability by Group", fontsize=14, weight="bold")
plt.xlabel("Group", fontsize=12)
plt.ylabel("Number of Recordings", fontsize=12)
plt.xticks(rotation=25, ha="right")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
df1 = df[0:6]
#dropping lgbt+ row
df1 = df1[df1['Group'] != 'LGBT+']
```

```python
# Pie chart for recording distribution by Racial Group
racial_group_counts = df1.groupby('Group')['Recordings Count'].sum()
plt.figure(figsize=(13, 6))
plt.title('Recording Distribution by Racial Group', fontsize=14, weight="bold")
plt.pie(racial_group_counts, labels=racial_group_counts.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.legend(title='Groups', loc='upper right')
plt.show()

```
